```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV
data = pd.read_csv('../data/expected_values_experiment.csv')

# Extract n and Expected_sqrt(n)
n = data['n']
expected_n_2 = data['Expected_n/2']
expected_sqrt_n = data['Expected_sqrt(n)']
```

```python
# Define the model function
def model_function1(n, c):
    return c

# Perform curve fitting
popt, pcov = curve_fit(model_function1, n, expected_n_2)

# Extract the best-fit value of c
best_fit_c = popt[0]
print("Best-fit value of c:", best_fit_c)
print("Expected_n/2:", best_fit_c)

# Plot the data and the fitted curve
plt.scatter(n, expected_n_2, label='Data')
# plt.plot(n, model_function1(n, best_fit_c), color='red', label='Fitted curve')
plt.xlabel('n')
plt.ylabel('Expected_n/2')
plt.title('Curve Fitting')
plt.legend()
plt.show()
```

```python
# Define the model function
def model_function(n, c):
    return c * np.sqrt(n)

# Perform curve fitting
popt, pcov = curve_fit(model_function, n, expected_sqrt_n)

# Extract the best-fit value of c
best_fit_c = popt[0]
print("Best-fit value of c:", best_fit_c)
print("Expected_sqrt(n):", best_fit_c, "* sqrt(n)")
# Plot the data and the fitted curve
plt.scatter(n, expected_sqrt_n, label='Data')
plt.plot(n, model_function(n, best_fit_c), color='red', label='Fitted curve')
plt.xlabel('n')
plt.ylabel('Expected_sqrt(n)')
plt.title('Curve Fitting')
plt.legend()
plt.show()
```
