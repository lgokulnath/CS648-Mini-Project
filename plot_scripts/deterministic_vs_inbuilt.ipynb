```python
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV
data = pd.read_csv('../data/deterministic_vs_inbuilt_times.csv')

# Extract n and Expected_sqrt(n)
n = data['n']
field1 = data['deterministic_time']
field2 = data['inbuilt_time']
```

```python
# Define the model function
def model_function(n, c):
    return c * n * n * n

# Perform curve fitting
popt, pcov = curve_fit(model_function, n, field1)

# Extract the best-fit value of c
best_fit_c = popt[0]
print("Best-fit value of c:", best_fit_c)
print("Time(deterministic):", best_fit_c, "* n^3 seconds")
# Plot the data and the fitted curve
plt.scatter(n, field1, label='Data')
plt.plot(n, model_function(n, best_fit_c), color='red', label='Fitted curve')
plt.xlabel('n')
plt.ylabel('Time(deterministic) in sec')
plt.title('Curve Fitting')
plt.legend()
plt.show()
```

```python
# Define the model function
def model_function(n, c):
    return c * n

# Perform curve fitting
popt, pcov = curve_fit(model_function, n, field2)

# Extract the best-fit value of c
best_fit_c = popt[0]
print("Best-fit value of c:", best_fit_c)
print("Time(inbuilt):", best_fit_c, "* n seconds")
# Plot the data and the fitted curve
plt.scatter(n, field2, label='Data')
plt.plot(n, model_function(n, best_fit_c), color='red', label='Fitted curve')
plt.xlabel('n')
plt.ylabel('Time(inbuilt) in sec')
plt.title('Curve Fitting')
plt.legend()
plt.show()
```
